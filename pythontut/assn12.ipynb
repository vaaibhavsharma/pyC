```python
import matplotlib.pyplot as plt 
import numpy as np
plt.style.use('ggplot') 
a = [0, 1, 2, 3]
b = [0, 10, 40, 30]
plt.plot(a, b, color='purple')
plt.xlabel('a')
plt.ylabel('b')
plt.title('simple plot')
plt.plot([10,5,20,35], color='black')
```

```python
x1 = [1,2,3,4,5,6]
y1 = [2,4,1,5,2,6]
plt.plot(x1, y1,'*--', color='purple', label='PurpleStars&DashedLine') # *(stars); --(dashed line);
plt.plot([1,2,3,4,5], [1,2,3,4,10], 'ko-', label='BlackDots&SolidLine') # k(black); o(dots); -(solid line)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
a = [1, 2, 3, 4, 5]
b = [10, 24, 36, 40, 5]
plt.bar(a, b, color='lavender', edgecolor='black')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('bar plot')
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 10)
y = np.exp(np.sin(x))
plt.stem(x, y,'k:', use_line_collection = True) # k(black); :(dotted line)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('stem plot')
plt.show()
```

```python
a = np.array([0, 2, 4, 6, 7])
b = np.array([1, 8, 16, 30, 50])
plt.step(a, b,'*-', where='post', color='purple')
plt.step(a, [10, 20, 25, 45, 60],'*-', where='pre', color='black')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('step plot')
plt.show()
```

```python
g=[4, 8, 14, 18, 20]
h=[20, 30, 40, 60, 80]
i=[30, 40, 60, 90, 90]
plt.fill_between(g, h, color='lavender', edgecolor='black')
plt.title('fill between plot')
plt.fill_between(g, h, i, color='grey', edgecolor='black')
plt.show()
```
