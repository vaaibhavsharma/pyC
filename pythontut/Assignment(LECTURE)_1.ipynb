# Regular Expressions ------------------------------

##### A Regular Expressions (RegEx) is a special sequence of characters that uses a search pattern to find a string or set of strings. It can detect the presence or absence of a text by matching with a particular pattern, and also can split a pattern into one or more sub-patterns.



```python
import re
  
s = 'Vaibhav Sharma 201334'
  
match = re.search(r'201334', s)
  
print('Start Index:', match.start())
print('End Index:', match.end())
```

#### re.compile() 
###### Regular expressions are compiled into pattern objects, which have methods for various operations such as searching for pattern matches or performing string substitutions. 



#### re.findall()
###### Return all non-overlapping matches of pattern in string, as a list of strings. The string is scanned left-to-right, and matches are returned in the order found.


#### re.split() 
###### Split string by the occurrences of a character or a pattern, upon finding that pattern, the remaining characters from the string are returned as part of the resulting list.

# Extracting email addresses using regular expressions in Python


```python
# SEARCHING IN DATA FILE
# FILE -------------------------------------------------------------------------------


# From jagpreet.sidhu\@juit.ac.in Sat Oct 23 09:14:16 2021
# Frm jagpreet.sidhu\@juit.ac.in Sat Oct 23 09:14:16 2021
# Fro jagpreet.sidhu\@juit.ac.in Sat Oct 23 09:14:16 2021
# From jagpreet.sidhu\@juit.ac.in Sat Oct 23 09:14:16 2021
# jagpreet.sidhu\@juit.ac.in Sat Oct 23 09:14:16 2021
# jagpreet.sidhu\@juit.ac.in Sat Oct 23 09:14:16 2021

# ------------------------------------------------------------------------------------


import re
f  = open('email.txt', 'r')
linesList = f.readlines()
listEmails = []
# Check if line Starting from "From"
for line in linesList:
    if line[:4] == "From":
        # \S matches any non-whitespace character 
        # @ for as in the Email 
        # + for Repeats a character one or more times 
        email = re.findall('\S+@\S+', line)
        listEmails.append(email)

print(listEmails)
```

```python
import re
  
# * = many no. of occurrence
p = re.compile('ok*')
print(p.findall("areuokgghjefjgokgshhgokokok"))
```
