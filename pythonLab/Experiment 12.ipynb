```python

def decoratorOne(func):
    def wrapper():
        print("1")
        func()
        print("3")
    return wrapper
 
 
def decoratorTwo(func):
    def wrapper():
        func()
        func()
        func()
    return wrapper
 
 
@decoratorOne
@decoratorTwo
def nice():
    print("2")
 
nice()
```
