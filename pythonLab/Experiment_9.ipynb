# TYPE CODES FOR PYTHON ARRAY 

# CommonArticleDesign2-min.png

```python
from array import *

val = array("i", [1,2,3,4,5])

print(val)
print(val[0])

print(list(val))

print(val.tolist())
print(val.typecode)
val.reverse()
print(val)

for i in val:
  print(i)

newarray = array(val.typecode, [i*i for i in val])

print(newarray)
```
